# Activation Function
* An Activation Function decides whether a neuron should be activated or not. This means that it will decide whether the neuron's input to the network is important or not in the process of prediction using simpler mathematical operations.
* The Activation function is categorized into two main parts:
    * Binary step function
    * Linear Activation function
    * Non-Linear Activation function

## Binary Step Function
* Binary step function depends on a threshold value that decides whether a neuron should be activated or not.




## Linear Activation Function
* The output of functions is not restricted in between any range. Its range is specified from -∞ (-infinity) to +∞ (+infinity).





## Non-Linear Activation Function
* These are one of the most widely used activation function. It helps the model in generalizing and adapting any sort of data in order to perform correct differentiation among the output.



* Non-linear activation functions
    * Sigmoid/Logistic Activation Function
    * Tanh Function (Hyperbolic Tongent)
    * ReLU Function
    * Softmax Function

# Loss Function
* The loss functin ia a method of evaluating how well your algorithm is modeling your dataset. It is a mathematical function of the parameters of the machine learning algorithm.
* types of loss functions in different deep learning models:
    * **Regression**:
        * MSE(Mean Squared Error)
        * MAE(Mean Absolute Error)
        * Hubber loss
    * **Classification**:
        * Binary cross-entropy

    * **Auto Encoder**:
        * KL Divergence
    * **GAN**:
        * Discriminator loss
        * Minmax GAN loss
    * **Object detection**:
        * Focal loss
    * **Word embeddings**:
        * Triplet loss
