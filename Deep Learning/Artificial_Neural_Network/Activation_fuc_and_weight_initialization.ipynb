# Activation Function
* **Sigmoid Activation Function**
    *  
    * 
    * Advantage:
        * output in the range of 0 and 1, so that it used mostly on the output layer and in classification problem.
        * Non-linear Function, so that it can capture non-linear data
        * it's equation is differentiable 
    * Disadvantage:
        * Saturating Function, which means it makes data in the range of 0 to 1 which can couse vanishing gradient problem.
        * Non-zero centered, which couse slow training.
        * it is computationally expensive, which means it's equation is very complex so that computer needs more power to computer it.
* **Tanh Activation Function** 
    *  
    * 
    * Advantage:
        * Non-linear Function, so that it can capture non-linear data as well.
        * it's equation is Differentiable
        * Zero-centered, so that it's gradient was postives as well as negetives that's why it's training is faster as compared to sigmoid.
    * Disadvantage:
        * Saturating Function, which cause to vanishing gradient problem.
        * Computationally expensive
* **Relu Activation Function**
    *  
    * 
    * Advantage:
        * Non-linear
        * Not Saturated in the positive region
        * Computationally inexpensive
        * Convergance is faster then sigmoid and tanh.
    * Disadvantage:
        * it is not completelly differentiable at zero.
        * it is not zero centered
        * dying relu problem
            * In it the some of the neuron's output is becomes zero and it does not depends on the input, and that neuron is called dead neuron. If any neuron become dead neuron then that neuron becomes dead for forever.
            * Reasons: 
                * High learning rate
                * High -ve bias
            * Solutions:
                * set low learning rate
                * set bias with +ve value
                * Don't use relu, use it's variants
    * **Varients of Relu**
        * Linear
            * Leaky Relu
            * Parametric Relu
        * Non-linear
            * Exponential linear unit (Elu)
            * Scaled Exponential linear unit (Selu)
