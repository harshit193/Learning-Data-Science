### 2. using scikitlearn OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
## it gives sparse matrix which means that the matrix contains only 0 and 1
## for converting it into the array we need to use .toarray() function
arr = ohe.fit_transform(gen_mar_col).toarray()

## converting it into the dataframe
pd.DataFrame(arr,columns=['Gender_female','Gender_Male','Married_No','Married_Yes'])
```

* But here the problem is that it create new two columns which increased the length of the dataset and adds some unwanted columns which is not required
* For solving this problem we are using drop='first' method in ohe = OneHotEncoder() it
* so the new code becomes ohe = OneHotEncoder(drop='first')
    * if we pass it gender then it creates two columns which has names as gender_male and gender_female but we don't want two columns then we are using drop first. the drop first drops the first columns and gives only second columns which is gender_female. Because this columns has two attributes which is 1 (true) or 0 (false) then we can easily identify that the number of males and females present in the dataset  

```python
ohe = OneHotEncoder(drop='first')
pd.DataFrame(ohe.fit_transform(gen_mar_col).toarray(),columns=['Gender_Male','Married_Yes'])
```

### Ordinal Encoding
* ordinal encoding prefered rather then label encoding
* it can be performed by two ways
    * cyclic line
    * map function

### Ordinal Encoding using cyclic line

```python
df = pd.DataFrame({'size':['s','m','l','xl','s','m','l','s','s','l','xl','m']})
df
```

```python
order = [['s','m','l','xl']]
```

```python
df['encoded_size'] = oe.transform(df[['size']])
df
```

### Outlier removal using IQR method

```python
df = pd.read_csv('./dataset/Loan.csv')
df.info()
df.head()
```

```python
df.describe()
```

```python
## iqr means inter quantile range
## for finding the quantile we use .quantile function
q1 = df['CoapplicantIncome'].quantile(0.25)
q3 = df['CoapplicantIncome'].quantile(0.75)
iqr = q3-q1  

min_range = q1 - (1.5*iqr)
max_range = q3 + (1.5*iqr)
min_range,max_range
```

```python
## because the min_range is negative value so we are discarding them and using only postivite value
new_df = df[df['CoapplicantIncome']<max_range]
```

```python
sns.boxplot(x=new_df['CoapplicantIncome'])
```

```python
df.describe()
sns.boxplot(x=df['CoapplicantIncome'])
plt.show()
sns.histplot(x=df['CoapplicantIncome'],kde=True)
plt.show()
```

#### Removing outlier using normal distribution graph
* minimum range = mean - 3 * stadard deviation
* maximum range = mean + 3 * stadard deviation

```python
min_range = df['CoapplicantIncome'].mean() - 3 * df['CoapplicantIncome'].std()
max_range = df['CoapplicantIncome'].mean() + 3 * df['CoapplicantIncome'].std()
min_range,max_range
```

```python
## becaue the min_range is in negative so we are only using max_range value
new_df = df[df['CoapplicantIncome']<=max_range]
new_df.shape
```

```python
sns.boxplot(x=new_df['CoapplicantIncome'])
plt.show()
```

### Removing outiler using z-score
> formula :- 
> z = ( x - mean ) / standard daviation

```python
z_score = (df['CoapplicantIncome'] - df['CoapplicantIncome'].mean())/df['CoapplicantIncome'].std()
df['z_score'] = z_score
df.shape
```

## Feature Scaling (Standardization)
* types of feature scaling
    1. Standardization
    2. Normalization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. Standardization
* It is a very effective technique which re-scales a feature value so that it has distribution with 0 mean value and variance equals to 1.
* Xnew = ( X - Xmean ) / standard Deviation

```python
df = pd.read_csv('./dataset/Loan.csv')
df.info()
df.describe()
```

* Scaling of the data with the help of sklearn library

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(df[['ApplicantIncome']])
```

```python
df['ApplicantIncome_ss'] = pd.DataFrame(ss.transform(df[['ApplicantIncome']]),columns=['ApplicantIncome_ss'])
df.head()
```

```python
df.describe()
```

#### 2. Normalization (Min-Max scaling)
* It is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.
* Xnew = ( X - min(X) ) / ( max(X) - min(X) )

```python
df = pd.read_csv('./dataset/Loan.csv')
df.describe()
```

```python
sns.histplot(df['CoapplicantIncome'],kde=True)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
## it works only on 2d data that's why we are using 2d array 
ms.fit(df[['CoapplicantIncome']])
```

```python
df['CoapplicantIncome_mms'] = pd.DataFrame(ms.transform(df[['CoapplicantIncome']]),columns=['CoapplicantIncome_mms'])
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Before Min-Max Scaling')
sns.histplot(df['CoapplicantIncome'],kde=True)
plt.subplot(1,2,2)
plt.title('After Min-Max Scaling')
sns.histplot(df['CoapplicantIncome_mms'],kde=True)
plt.show()
```

# Model Parameter
* Model Parameter are configuration variables that are internal to the model, and a model learns them on its own.
* for example
    * in linear regression the linear line formula(y=mx+c) is used for finding the best fit line, in this formula the **m** and **c** is called model parameter because this parameters are helping the model for predicting the output

    

# Hyper Parameter
* Hyper parameters are those parameters that are explicitly defined by the user to control the learning process.
* The best value can be determined either by the rule of thumb or by trial and error.
* for example
    * in decision tree there many other parameters mostly which is a default values (like max_depth, max_features, splitter, max_leaf_nodes, presort etc.) are used for controling the model output prediction which helps to improve the accuracy of the model for predicting the more accurate output, this parameter are called **Hyper Parameter**.

    

# Hyper Parameter Tuning
* Models can have many hyperparameters and finding the best combinatin of parameters can be treated as a search problem. The two best strategies for Hyperparameter tuning are:
    * GridSearchCV
    * RandomizedSearchCV

## 1. GridSearchCV
* GridSearchCV is a technique to search through the best parameter values from the given set of the grid of parameters.

## 2. RandomizedSearchCV
* It goes through only a fixed number of hyperparameter settings.
* It moves within the grid in a random fashion to find the best set of hyperparameters.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('../dataset/polynomial.csv')
df.info()
df.head()
```

```python
sns.scatterplot(x=df['Level'],y=df['Salary'])
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df['Salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
dt = DecisionTreeRegressor(criterion='squared_error',max_depth=9,splitter='best')
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train),dt.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
Hyper_Parameters = {
    "criterion":['squared_error','friedman_mse','absolute_error','poisson'],
    "splitter":['best','random'],
    "max_depth":[i for i in range(2,21)]
}
```

```python
gd = GridSearchCV(DecisionTreeRegressor(),param_grid=Hyper_Parameters)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_
```

# Cross Validation 
* Cross-validatin is a technique for validating the model efficiency by training it on the subset of input data and testing on previously unseen subset of the input data.



* Methods used for cross-validation
    * Leave p out cross-validation
    * Leave one out cross-validation
    * Holdout cross-validation
    * Repeated random subsampling validation
    * k-fold cross-validation
    * Stratified k-fold cross-validation
    * Time Series cross-validation
    * Nested cross-validation

## K-fold cross-validation
* The original dataset is equally partitioned into k subparts or folds. Out of k-folds or groups, for each iteration, one group is selected as validation data, and the remaining (k-1) groups are selected as training data. Not suitable for an imbalanced dataset.

## Stratified K-fold cross-validation
* The original dataset is equally partitioned into k subparts or folds. Out of the k-folds or groups, for each iteration, one group is selected as validation data, and the remaining (k-1) groups are selected as training data.
* Stratified k-fold cross-validation solved the problem of an imbalaned dataset.

## Leave one out cross-validation
* It is an exhaustive cross-validation technique. It is a category of Lp OCV with the case of p=1.



## Leave p out cross-validation
* This is same as leave one out cross-validation but in it we can select the combinations of the data which is used for validation
