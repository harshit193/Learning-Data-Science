# Association Rule
* Association rule learning is a machine learning technique used to discover intersting relationships or patterns in large datasets.
* It is often applied to transactional data, where items are bought or used together.
* Association rule learning works on the concept of if and else statement, such as if A then B.



* If element is called antecedent, and then statement is called as consequent.
* it works on 
    * Support 
    * Confidence
    * Lift
* **Support**: The first step for us and the algorithm is to find frequently bought items. It is a straight forward consdition that is based on frequency.
> 
* **Confidence**: We have identified frequently bought items let's calculate confidence. This will tell us how confident (based on our data) we can be that an item will be purchased, given that another item has been purchased.
> 
* **Lift**: Given that different items are bought at different frequencies (strong association)
> 
* If:
    * **lift>1**: means that the two items are more likely to be bought together.
    * **lift<1**: means that the two items are more likely to be bought separatly.
    * **lift==1**: means that there is no association between the two items.
* Algorithms which works on Association rule learning
    * Apriori
    * Eclat
    * F-P Growth Algorithm

    * Market Basket Analysis
