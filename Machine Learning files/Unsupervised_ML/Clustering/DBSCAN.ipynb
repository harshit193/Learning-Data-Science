# DBSCAN (Density-Based Spatial Clustering of Application with Noise) Clustering
* The clusters found by DBScan can be any shape, which can deal with some special cases that other methods cannot.
* It can be used for non-linear dataset 
* It also used for outlier detection because it is robust with outliers that's why it can easily identify the outliers and indicates them separately.
* Difference between K-means and DBSCAN Clustering





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=250, noise=0.05)
print(x,y)
```

```python
df = pd.DataFrame(x,columns=['data1','data2'])
df['output'] = y
df.info()
df.head()
```

```python
sns.scatterplot(x=df['data1'],y=df['data2'],hue=df['output'])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2,min_samples=5)
db.fit_predict(df.drop(columns='output'))
```

```python
df['predict'] = db.fit_predict(df.drop(columns='output'))
df.head()
```

```python
sns.scatterplot(x=df['data1'],y=df['data2'],hue=df['predict'])
plt.show()
```
