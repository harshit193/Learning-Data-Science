   * formula : ∑|xi-mean|/n
   * here  xi means the ith element of the x columns and the mean of that column, n means the total number of elements present in the column x

* abs referse as absolute value it gives only positive value

```python
#formula for finding the mean absolute deviation
# mean_absolute_deviation_sec_a = np.abs(sec_a-mean).sum()/len(sec_a)
# mean_absolute_deviation_sec_b = np.abs(sec_b-mean).sum()/len(sec_b)

mean_absolute_deviation_sec_a = np.sum(np.abs(sec_a-mean))/len(sec_a)
mean_absolute_deviation_sec_b = np.sum(np.abs(sec_b-mean))/len(sec_b)

mean_absolute_deviation_sec_a,mean_absolute_deviation_sec_b
```

# Skewness
Skewness is devided into three parts:-
1. Negatively skewed data 
    * mean < median < mode
2. Normal (no skew) data
    * mean = median = mode
3. Positively skewed data
    * mode < median < mean

consider the values of mean, median and mode of the x axis

```python
print('the skewness of the data is ', df['age'].skew())
mean = df['age'].mean()
median = df['age'].median()
mode = df['age'].mode()[0]
print(f'mean : {mean}, median : {median}, mode : {mode}')
sns.histplot(df['age'],kde=True,bins=[i for i in range(0,81,5)])
plt.plot([mean for i in range(0,100)],[i for i in range(0,100)],color = 'r',label='mean')
plt.plot([median for i in range(0,100)],[i for i in range(0,100)],color = 'b',label='median')
plt.plot([mode for i in range(0,100)],[i for i in range(0,100)],color = 'purple',label='mode')
plt.legend()
plt.show()

## the following graph shows positive skewness because the mode < median < mean 
```

## Central Limit Theorem
* central limit theorem is used for converting non normal distribution data into normal distribution data
* for converting into normal distribution data we need to follow some rules
    * The sample data must be sampled and selected randomly from the population.
    * There should not be any multicollinearity in the sampled data which is one sample should not influence the other samples.
    * The sample size should be no more than 10% of the population or the actual data. Generally the sample size greater than 30 (n>30) is considered for good result.

```python
pop_data = np.random.randint(10,100,10000)
pop_table = pd.DataFrame({'pop_data':pop_data})

plt.figure(figsize=(3,2))
sns.kdeplot(x='pop_data',data=pop_table)
plt.show()
```

### Here we have the non normal distributed data of the size of 10000 
* for converting it into the normal distributed data we need to take the random samples of the size of less then 10% of the data
    * sample data size is :- (10000*10)/100 = 1000
    * we need to take the sample data of the length of less then 1000
* then we need to find the mean value of that data for converting it into the normal distribution

```python
sample_mean = []
for i in range(50):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
```

```python
plt.figure(figsize=(3,2))
sample_M = pd.DataFrame({'sample_mean':sample_mean})
sns.kdeplot(sample_M['sample_mean'])
plt.show()
```

```python
print(np.mean(pop_data),np.mean(sample_M['sample_mean']))
```

## Hypothesis Testing
* State null ( H0 ) and alternative ( H1 ) hypothesis
* choose lever of significance ( α )
* Find critical values
* Find test statistic
* Draw your conslusion

### There are three methods of checking the Hypothesis testing
#### Z-test
* we need the mean of the population data
* standard deviation of the population data
* the length of the sample must me grater then 30
* and the mean of the sample data 
* Formula:- 
    > Z-test = (sample mean - population mean)/(populatin standard deviation/root of number of samples)

#### T-test
* mean of the population data
* standard deviation of the sample data
* the length of the sample is less then 30
* mean of the sample data
* Formula:- 
    > T-test = (sample mean - population mean) / (sample standard divation/root of number of samples)

#### chi-square test
* It is applicable in two types of situations
    * Goodness (it is used for finding that the data is correct or not)
    * Independence (it checks that the two data's are dependent on each other or not)
* Formula:- 
    > X² = ∑ ( ( Oⅰ - Eⅰ )² / Eⅰ )
    >> Here 
    >>> X² = chi squared
    >>> Oⅰ = observed value
    >>> Eⅰ = expected value
