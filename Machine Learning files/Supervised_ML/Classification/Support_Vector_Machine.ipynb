# Support Vector Machine (in Classification)
* SVM in one of the most popular Supervised Machine Learning algorithm, which is used for Classification as well as Regression problems.


* **Hard Margin**: The algorithm aims to find a hyperplane that perfectly separates the data into two classes without any misclassifications.
* **Soft Margin**: the algorithm allows for some misclassifications to find a hyperplane that generalizes better to unseen data and is more robust to outliers.
* Types of SVMs:
    * **Simple SVM**: Typically used for linear regression and classification problems.
    * **Kernel SVM**: Has more flexibility for non-linear data because you can add more features to fit a hyperplane instead of a two-dimentional space.
* **Kernel Functions**: 
    * Kernel functions play a crucial role in tranforming input data into a higher-dimensional space.
    * The primary purpose of kernel functions is to allow SVMs to handle non-linearly seprable data by implicitly mapping the input data into a higher-dimensinal feature space where linear separation may be more feasible.
    * This transformation is done without explicitly calculating the coordinates pints in that higher-dimensional space.
* Types of kernels




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
```

```python
df = pd.read_csv('../../../dataset/Social_Network_Ads.csv')
df.info()
df.head()
```

```python
df.drop(columns=['User ID','Gender'],inplace=True)
df.rename(columns={'EstimatedSalary':'Salary'},inplace=True)
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
sns.scatterplot(x=df['Age'],y=df['Salary'],hue=df['Purchased'])
plt.show()
```

```python
x = df.drop(columns='Purchased')
y = df['Purchased']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

### in rbf kernel

```python
svc = SVC(kernel='rbf')
svc.fit(x_train, y_train)
```

```python
svc.score(x_train,y_train), svc.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svc)
plt.show()
```

### in linear kernel

```python
svc = SVC(kernel='linear')
svc.fit(x_train,y_train)
svc.score(x_train,y_train),svc.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svc)
```

```python
svc = SVC(kernel='poly')
svc.fit(x_train,y_train)
svc.score(x_train,y_train),svc.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svc)
plt.show()
```

```python
svc = SVC(kernel='sigmoid')
svc.fit(x_train,y_train)
svc.score(x_train,y_train),svc.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svc)
plt.show()
```
