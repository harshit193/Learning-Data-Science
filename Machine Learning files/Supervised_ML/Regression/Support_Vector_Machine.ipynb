# Support Vector Machine (in Regression)
* Support vector machine is a regression technique that uses Support Vector Machine for modeling and predicting continuous outcomes.




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('../../../dataset/placement.csv')
df.info()
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['package'])
plt.show()
```

```python
x = df[['cgpa']]
y = df['package']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
svr = SVR(kernel='rbf')
svr.fit(x_train,y_train)
```

```python
svr.score(x_train,y_train),svr.score(x_test,y_test)
```

```python
sns.scatterplot(x=df['cgpa'],y=df['package'])
plt.scatter(df['cgpa'],svr.predict(x),color='r')
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy().astype(np.int64),clf=svr)
plt.show()
```
