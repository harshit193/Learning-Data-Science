# Linear Regression Algorithm

## 1. Simple Linear Regression Algorithm
* Simple Linear Regression is a type of Regressuin algorithms that models the relationship between a dependent variable and a sinlge independent variable.
* It is used when the input is a single value or which means we have only one input feature
* Linear Equation :- y = mx + c
* where: 
    * y = dependent variable
    * x = independent variable 
    * m = slope/gradient/coefficient and m = (x2 - x1) / (y2 - y1)
    * c = intercept
* here the value of m denotes the angle formed by line with respect to x-axis
    * if the value of m is positive then the angle formed by line is between 0 and 90 ( 0<∅<90 )
    * if the value of m is negative then the angle formed by line is greater then 90 ( ∅>90 )
    * if the value of m is equal to 0 then the angle is 0 ( ∅=0 )
* here the value of the c denotes that the interception point of the line with y-axis
    * if it is positive then the line is intercepting into the positive part of the y-axis
    * if it is negative then the line is intercepting into the negative part of the y-axis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('../../dataset/placement.csv')
df.info()
df.head()
```

```python
df.isnull().sum()
```

* In machine learning we need to give the data into the two dimentional because it only takes two dimentional or more then two dimentional data for predicting the ouput
* If we are using the one dimentional data then it gives error
* and if we are using the sklearn library then we need to provide it also a two dimentional data 
> Note: it is not compulsory to pass the two dimentinal data into second number of value

* x: The input features (independent variables), usually a 2D array or DataFrame.

* y: The labels/targets (dependent variable), usually a 1D array or Series.



```python
x = df[['cgpa']]
y = df[['package']]

## ndim is used for finding the dimention of the data
print(f'dimention of x : {x.ndim} and dimention of y : {y.ndim}')
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x=df['cgpa'],y=df['package'])
plt.show()
```

```python
## in it we are giving the the size of the train data to split
## we also can be give the test_size of the data for testing 
## 
# x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.75, random_state=50)
```

* here the LinearRegression is a class and we are declaring the object of that class for using the functionality of that class

```python
lr = LinearRegression()
```

* here the lr.fit is finding the best value of the m and c of the linear regression formula so that the line is draw from the center of the data

```python

lr.fit(x_train,y_train)
```

* for predicting the output from that model we are using the lr.predict method for predicting the output according to the train data
* this method takes two dimentional data and gives the predicted output value
* Working of predict method
(method) def predict(X: MatrixLike) -> ndarray
Predict using the linear model.

Parameters
X : array-like or sparse matrix, shape (n_samples, n_features)
Samples.

Returns
C : array, shape (n_samples,)
Returns predicted values.

```python
# lr.predict([[9.9]])
lr.predict(x)
```

* the .score() method is used to evaluate how well the model fits the data.

```python
lr.score(x_test,y_test)
```

* In the formula y = mx + c 
    * here m is a coeficiant and to find it we are using lr.coef_ 
    * here the value of m = 0.56826736 which is positive which means that the angle of the line is between 0 and 90

```python
lr.coef_
```

* In the formula y = mx + c
    * here c is a intercept and to find it we use lr.intercept_ 
    * here the value of c = -0.99153783 which is in negative which means that the line is formed into the negative y axis

```python
lr.intercept_
```

* so the final formula of the linear regression is y = mx + c
    * y = 0.67826736*x + (-0.99153783)

```python
lr.predict([[6.89]])
```

```python
0.56826736 * 6.89 -0.99153783
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['cgpa'],y=df['package'])
plt.plot(df['cgpa'],lr.predict(x),color='r')
plt.legend(['training data','predicted data'])
plt.show()
```

## 2. Multiple Linear Regression Algorithm
* Multiple Linear Regression is an extension of simple linear regression as it takes more than one predictor variable to predict the response variable.
* this is used when the number of inputs is more then one.
* formula :- y = m1x1 + m2x2 + m3x3 + ...... + c


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../../dataset/salary-data.csv')
df.info()
df.head()
```

```python
df.rename(columns={'Years of Experience':'Experience'},inplace=True)
```

```python
df.drop(columns=['Education Level','Gender','Job Title'],inplace=True)
df.info()
df.head()
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    df.fillna({i:df[i].mean()},inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
x = df.drop(columns='Salary')
x.shape
```

```python
y = df['Salary']
y.shape
```

```python
df.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.75, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
